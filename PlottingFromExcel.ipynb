```python
#Import modules 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math 
from scipy.optimize import curve_fit
```

```python
#Import data from excel - in this case, we plot a thixotropy test (time vs. modulus) for collagen 

#Import the excel file, from location on desktop with appropriate sheet number 
file = pd.read_excel(r"/Users/naomiduggan/Desktop/Internship/Data/PPData.xlsx", sheet_name="NoCells") 

#Call on specific columns on the sheet 
time = file['169S4Time'] #Time 
SM = file['169S4SM'] #Storage modulus 
LM = file['169S4LM'] #Loss modulus 
```

```python
#Plotting 

fig, ax = plt.subplots()

#Plotting SM and LM 
#Plot for x (time) and y (SM and LM)
#Marker can be set to ["." , "," , "o" , "v" , "^" , "<", ">"]
#Label for legend 
#Color can be set to any HTML color code (https://htmlcolorcodes.com/) 
ax.plot(time, SM, '^', label='SM', color = '#425E8A') 
ax.plot(time, LM, 'o', label='LM', color = '#7493C3') 

#Grid line
ax.grid(which = "major", linewidth = 1)
ax.grid(which = "minor", linewidth = 0.2)
ax.minorticks_on()

ax.set_xscale('log') #Set to log scale 
ax.set_yscale('log') #Set to log scale 

#Labels
plt.xlabel("Time [min]") 
plt.ylabel("Modulus [Pa]") 
plt.title('Thixotropy Test for Collagen')

#Legend, can change location by altering parameters 
ax.legend(loc='upper left', bbox_to_anchor=(0.3, 1), ncol=3, fancybox=True, shadow=True)

plt.ylim(0.01, 100)

plt.show()
```
